```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('/Users/fortuneudeh/Desktop/projectTwo/neighborhood-childcare-affordabilty-analysis.csv', index_col='Neighborhoods')
# Convert MedianIncome to numeric by removing non-numeric characters (like $ and ,) and sorting the data
data['MedianIncome'] = data['MedianIncome'].replace('[\$,]', '', regex=True).astype(float)
sorted_data = data.sort_values(by='MedianIncome')

# Plot the sorted data
sns.scatterplot(x=sorted_data.index, y='MedianIncome', data=sorted_data)
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
```
